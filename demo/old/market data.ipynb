# Market Data

Se muestra la tabla del precio de cada minuto del S&P 500.

```python
import psycopg2
import pandas as pd
```

```python
hostname = 'localhost'
database = 'airflow'
username = 'airflow'
pwd = 'airflow'
port_id = 5433

# Conn es un objeto de conexión a la base de datos
try:
    conn = psycopg2.connect(
        host=hostname,
        dbname=database,
        user=username,
        password=pwd,
        port=port_id
    )
    print("Conexión exitosa")

    QUERYGSPC  = "SELECT * FROM market_data.gspc g ORDER BY g.ts DESC;"
    QUERYSP500 = "SELECT * FROM market_data.sp500 g ORDER BY g.ts DESC;"
    
    gspc  = pd.read_sql(QUERYGSPC , conn)
    sp500 = pd.read_sql(QUERYSP500, conn)

    # onn.close() # If you dont close the connection, it could lead to a connection leak. 
    # A connection leak occurs when database connections are not properly closed after their use, leading to exhaustion of available connections. This can result in performance degradation, application errors, and even system crashes as new connection requests are denied due to the lack of available connections. 
    # For example, in a web application, if each user request opens a database connection but fails to close it after the request is completed, over time, the number of open connections can accumulate, eventually reaching the maximum limit set by the database server. This can prevent new users from accessing the application and lead to downtime.
    print("Conexión cerrada")
except Exception as e:
    print(f"Ocurrió un error: {e}")
```
