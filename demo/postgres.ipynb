# Postgres to Python

Se muestra como conectarse a la base de datos Postgres desde Python y ejecutar consultas SQL.

```python
import psycopg2
```

```python
hostname = 'localhost'
database = 'airflow'
username = 'airflow'
pwd = 'airflow'
port_id = 5432

# Conn es un objeto de conexión a la base de datos
try:
    conn = psycopg2.connect(
        host=hostname,
        dbname=database,
        user=username,
        password=pwd,
        port=port_id
    )
    print("Conexión exitosa")

    conn.close()
    print("Conexión cerrada")
except Exception as e:
    print(f"Ocurrió un error: {e}")
```

```python
conn = psycopg2.connect(
    host=hostname,
    dbname=database,
    user=username,
    password=pwd,
    port=port_id
)
print("Conexión exitosa")

cur = conn.cursor()
print("Cursor creado")
```

```python
cur.execute("select * from market_data.gspc g order by g.ts desc;")
rows = cur.fetchall()
for row in rows:
    print(row)
```
